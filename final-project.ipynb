```python
import pandas as pd
import numpy as np
import numpy.matlib
from math import sqrt
import os
from numpy.linalg import matrix_power ## jic matrix mult introduce errors
from scipy.linalg import eig
import matplotlib.pyplot as plt
plt.rc('xtick', labelsize=16) 
plt.rc('ytick', labelsize=16) 

π = np.pi
np.set_printoptions(precision=3)

def createPlot(title="", x="k", y="Gr"):
    fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, sharex=True, sharey=True, figsize=(3*4, 2*4),dpi=300)
    plt.suptitle(title, fontsize=36, y=.95)
    fig.text(0.5, 0.04, x, ha='center', fontsize=28)
    fig.text(0.00, 0.5, y, va='center', rotation='vertical', fontsize=28)
    axes = [ax1,ax2,ax3,ax4,ax5, ax6]
    k = 0
    
    for ax in axes:
        ax.grid(b=True, which='major', linestyle='--', linewidth=.5*2)
        ax.grid(b=True, which='minor', linestyle='--', linewidth=.1*2)
        ax.minorticks_on()
    for ax in axes:
        for tick in ax.xaxis.get_major_ticks():
            tick.tick1line.set_visible(False)
            tick.tick2line.set_visible(False)
        for tick in ax.xaxis.get_minor_ticks():
            tick.tick1line.set_visible(False)
            tick.tick2line.set_visible(False)
        for tick in ax.yaxis.get_major_ticks():
            tick.tick1line.set_visible(False)
            tick.tick2line.set_visible(False)
        for tick in ax.yaxis.get_minor_ticks():
            tick.tick1line.set_visible(False)
            tick.tick2line.set_visible(False)
    plt.subplots_adjust(wspace=0.01,hspace=0.01)
    
    return fig, ((ax1, ax2, ax3), (ax4, ax5, ax6))

def createPlotArray(title="", x="", y="", M=4, N=8, lim=100, ytitle=.95):
    fig, axs = plt.subplots(M, N, sharex=True, sharey=True, figsize=(N*4, M*4))
    plt.suptitle(title, fontsize=36, y=ytitle)
    fig.text(0.5, 0.00, x, ha='center', fontsize=28)
    fig.text(0.05, 0.5, y, va='center', rotation='vertical', fontsize=28)
    for ax in axs.reshape(-1):
        ax.grid(b=True, which='major', linestyle='--', linewidth=.5*2)
        ax.grid(b=True, which='minor', linestyle='--', linewidth=.1*2)
        ax.minorticks_on()
        ax.axvspan(0, 2*lim, facecolor='r', alpha=0.05)
        for tick in ax.xaxis.get_major_ticks():
            tick.tick1line.set_visible(False)
            tick.tick2line.set_visible(False)
        for tick in ax.xaxis.get_minor_ticks():
            tick.tick1line.set_visible(False)
            tick.tick2line.set_visible(False)
        for tick in ax.yaxis.get_major_ticks():
            tick.tick1line.set_visible(False)
            tick.tick2line.set_visible(False)
        for tick in ax.yaxis.get_minor_ticks():
            tick.tick1line.set_visible(False)
            tick.tick2line.set_visible(False)
    plt.subplots_adjust(wspace=0.01,hspace=0.01)
    plt.xlim([-lim,lim])
    plt.ylim([-lim,lim])
    
    
    return fig, axs

def getBnd(NL,NGR,unstableGlobal,LL,GGR):

    ## north
    bndN = np.zeros( (NL-1,NGR), dtype=bool )
    bndN = np.logical_xor(unstableGlobal[:,:-1], unstableGlobal[:,1:])
    LLN = .5*(LL[:-1,:]+LL[1:,:])
    GGRN = .5*(GGR[:-1,:]+GGR[1:,:])

    ## east
    bndE = np.zeros( (NL,NGR-1), dtype=bool )
    bndE = np.logical_xor(unstableGlobal[:-1,:], unstableGlobal[1:,:])
    LLE = .5*(LL[:,:-1]+LL[:,1:])
    GGRE = .5*(GGR[:,:-1]+GGR[:,1:])

    ## northEast
    bndNE = np.zeros( (NL-1,NGR-1), dtype=bool )
    bndNE = np.logical_xor(unstableGlobal[:-1,:-1], unstableGlobal[1:,1:])
    LLNE = .5*(LL[:-1,:-1]+LL[1:,1:])
    GGRNE = .5*(GGR[:-1,:-1]+GGR[1:,1:])

    ## southEast
    bndSE = np.zeros( (NL-1,NGR-1), dtype=bool )
    bndSE = np.logical_xor(unstableGlobal[:-1,1:], unstableGlobal[1:,:-1])
    LLSE = .5*(LL[:-1,1:]+LL[1:,:-1])
    GGRSE = .5*(GGR[:-1,1:]+GGR[1:,:-1])

    X = []
    Y = []
    
    X.append( LLN[bndN.T==True] ) ; Y.append( GGRN[bndN.T==True] )
    X.append( LLE[bndE.T==True] ) ; Y.append( GGRE[bndE.T==True] )
    X.append( LLNE[bndNE.T==True] ) ; Y.append( GGRNE[bndNE.T==True] )
    X.append( LLSE[bndSE.T==True] ) ; Y.append( GGRSE[bndSE.T==True] )
    X = np.concatenate(X)
    Y = np.concatenate(Y)
    
    return X, Y

def createPlot3D( xlabel=r'k', ylabel=r'l', zlabel=r'Gs', title="", azim=-60, elev=30):
    fig, ax = plt.subplots(figsize=(16, 12), subplot_kw={"projection": "3d"})
    
    # make the panes transparent
    ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    
    # make the grid lines transparent
    ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
    ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
    ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)
    
    # adjust the camera angle
    ax.view_init(elev, azim)
    
    # adjust title, x & y labels, etc
    plt.title(title, fontsize=30)
    plt.xlabel(xlabel=xlabel, fontsize=24 ); plt.ylabel(ylabel=ylabel, fontsize=24 );# plt.zlabel(zlabel, fontsize=24 )
#     plt.xticks(fontsize=20 ); plt.yticks(fontsize=20 ); ax.set_zticks(fontsize=20 )
    ax.set_zlabel('GS',fontsize=24)
    ax.tick_params(axis='both', which='major', labelsize=20)
    ax.tick_params(axis='both', which='minor', labelsize=8)
    plt.grid(b=True, which='major', linestyle='--', linewidth=.5); plt.grid(b=True, which='minor', linestyle='--', linewidth=.1)
    plt.minorticks_on()
    
    return fig, ax





```

```python
def cheb(N):
    if N==0:
        D=0.
        x=1.
        return D,x
    else:
        z_n = np.arange(0.,N+1)
        x = np.cos(π*z_n/N); 
        c = (np.concatenate([[2.],np.ones(N-1),[2.]]))*((-np.ones(N+1))**z_n)
        X = np.matlib.repmat(x.T,N+1,1)
        dX = X.T-X
        D = (np.outer(c,(1./c)))/(dX+np.eye(N+1))
        D = D-np.diag(np.sum(D,axis=1))
        return D,x
    
```

```python
def assembleThreeDimensionalMode( N, l, k, Gs, Gr, Pr, Le ):
    kabs = sqrt(l**2 + k**2)
    ## assemble differentiation matrices D2, D4, identity matrix I and zero matrix O
    D,z = cheb(N)
    D2 = matrix_power(D,2);    D2 = D2[1:-1,1:-1] ## to apply B.C.s
    S0 = np.concatenate([[0.],1./(1.-z[1:-1]**2),[0.]]);           S = np.diag(S0)
    D4 = np.matmul(np.diag(1-z**2),matrix_power(D,4))  \
         -8.*np.matmul( np.diag(z),matrix_power(D,3))  \
         -12.*matrix_power(D,2)                         
    D4 = np.matmul(D4,S);      D4 = D4[1:-1,1:-1] ## to apply B.C.s
    D = D[1:-1,1:-1] ## to apply B.C.s
    D = 2.*D;
    D2 = 4.*D2;                D4 = 16.*D4 ## reminder that we are working now on [-.5,.5], not [-1.,+1.]
    
    I = np.eye(N-1);           O = np.zeros([N-1,N-1])

    ## reminder that we need to ignore the boundary points, and we are working now on [-.5,+.5]
    z = .5*z[1:-1]
    c0 = (1.-Gs/Gr)/6.
    Ub = c0 * (z**3-.25*z);                                        Ub = np.diag(Ub)
    dUb_dz = c0 * (3.*z**2-.25);                                   dUb_dz = np.diag(dUb_dz)
    d2Ub_dz2 = c0 * (6.*z);                                        d2Ub_dz2 = np.diag(d2Ub_dz2)
    dTb_dz = (Gr*Pr)*c0*(.25*z**4-.125*z**2+7./960.);              dTb_dz = np.diag(dTb_dz)
    dSb_dz = (Le*Gr*Pr)*c0*(.25*z**4-.125*z**2+7./960.);           dSb_dz = np.diag(dSb_dz)

    ## define the operators A, B
    A11 = D4-2*kabs**2*D2+kabs**4*I - 1j*l*Gr*(np.matmul(Ub,D2-kabs**2*I) - d2Ub_dz2); 
    A12 = O;           A13 = -kabs**2*I;                 A14 = (Gs/Gr)*kabs**2*I
    A21 = np.matmul(D2-kabs**2*I, -1j*l*D)-l**2*Gr*np.matmul(Ub,D)+k**2*Gr*dUb_dz
    A22 = (D2-kabs**2*I)*kabs**2 - 1j*l*kabs**2*Gr*dUb_dz
    A23 = O;                          A24 = O          
    A31 = -Gr*Pr*dTb_dz;       A32 = -Gr*Pr*I;    A33 = D2-kabs**2*I-1j*Gr*Pr*l*Ub; A34 = O
    A41 = -Le*Gr*Pr*dSb_dz;    A42 = -Le*Gr*Pr*I; A43 = O;         A44 = D2-kabs**2*I - 1j*Le*Gr*Pr*l*Ub
    A = np.block( [
            [A11, A12, A13, A14],
            [A21, A22, A23, A24],
            [A31, A32, A33, A34],
            [A41, A42, A43, A44]
        ]) ## should remove the redundant sub-blocks later (maybe not, for readability)

    B11 = D2-kabs**2*I;        B22 = kabs**2*I;   B33 = Pr*I;      B44 = Le*Pr*I
    B21 = -1j*l*D
    B = np.block([
            [B11, O,   O,   O  ],
            [B21, B22, O,   O  ],
            [O,   O,   B33, O  ],
            [O,   O,   O,   B44]
        ])

    return A,B

```
